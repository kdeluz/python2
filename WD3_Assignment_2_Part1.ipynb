# .unique()
### Shows all unique values in a column. It can be applied to a single column only, not the whole dataframe.
### eg. dataname['Column name'].unique()

# .nunique()
### shows the unique values in each column. Can be applied to both single columns and the whole dataframe.


```python
# This displays unique values in the snow column, and returns a numpy array containing all the unique values within the series.
# Cannot include anything within the bracket.
data['snow'].unique()
```

```python
# This instead finds the number of unique values in the SolarEnergy column.
# This method returns an integer that represents the count of unique values within the series.
data['solarenergy'].nunique()
```

# .count()
### Shows the total number of non-null values in each column.


# .value_counts()

```python
# Show sthe number of non-nnull values in each column.
# Can use without giving a column name, to see every column's count.
# data.count()
data['dew'].count()
```
